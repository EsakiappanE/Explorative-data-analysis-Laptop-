# OUTLIER DETECTION AND REMOVAL USING IQR

```python
df
```

```python
df.shape
```

```python
sns.boxplot(x=df['Price'])
```

```python
df1=df.select_dtypes(exclude=['object'])
df1
```

```python
q1=df1.quantile(0.25)
q1
```

```python
q3=df1.quantile(0.75)
q3
```

```python
IQR=q3-q1
IQR
```

```python
e=((df1<q1-1.5*IQR)|(df1>q3+1.5*IQR))
e
```

```python
e.sum()   
```

```python
filter=df[(~e).any(axis=1)]
```

```python
filter
```

```python
filter.shape
```
